```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create List
x = np.logspace(0,15,100,dtype=np.float128)
```

```python
#Create functions

#1
y_1 = np.ones(100) 

#n
y_n = x

#n^2
y_square = np.power(x,2)

#sqrt(n)
y_squareroot = np.sqrt(x)

#nlogn
y_nlogn = x*np.log(x)

#n^3
y_cube = np.power(x,3)

#2^n
y_2n = np.power(2,x)
```

```python
#Setting the plot to logscale
fig= plt.figure()
p = fig.add_subplot(1,1,1)
p.set_yscale('log')
p.set_xscale('log')
p.set_xlim(1,10**15)
p.set_ylim(1,10**19)


#Plotting the functions to the graph
p.plot(x,y_1,label='1')
p.plot(x,y_n,label='n')
p.plot(x,y_square,label='n^2')
p.plot(x,y_squareroot,label='sqrt(n)')
p.plot(x,y_nlogn,label='nlogn')
p.plot(x,y_cube,label ='n^3')
p.plot(x,y_2n,label='2^n')

p.set_xlabel("Input Size")
p.set_ylabel("Running Time Complexity")
p.legend()
```
